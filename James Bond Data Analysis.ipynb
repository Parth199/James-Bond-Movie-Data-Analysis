# Apply Method

```python
def convert_tostring_and_add_millions(number):
    return str(number) + "MILLIONS"
```

```python
bond.head(3)
```

```python
column=["Box Office","Budget","Bond Actor Salary"]
for col in column:
    bond[col]=bond[col].apply(convert_tostring_and_add_millions)
```

```python
bond.head(3)
```
